```
# Criação de novas colunas (a partir de colunas existentes )
# criando nova coluna a partir de game_event_id que será o valor original de 50
df["game_event_id_somando"]= df["game_event_id"] + 50  
```

```
df
```

```
def is_home_game(valorCadaLinha): # valor de cada linha da coluna matchup
    if valorCadaLinha.find('@') < 0:
        return 1
    else:
        return 0
    
#aplicar a função em cada valor da coluna matchup
df["home_game"] = df["matchup"].apply(is_home_game)
```

```
df
```

```
#usando apply com lambda 
df["home_game_lambda"] = df["matchup"].apply(lambda valorCadaLinha: 1 if (valorCadaLinha.find("@") < 0) else 0 )
```

```
# a partir de game_date, inferir ano, mes, dia dia da semana, e dia do ano
df["game_date"] = pd.to_datetime(df["game_date"]) # sobrescrevendo a coluna original após conversão para tipo de data
df['ano'] = df['game_date'].dt.year
df['mes'] = df['game_date'].dt.month
df['dia'] = df['game_date'].dt.day
df['dia_da_semana'] = df['game_date'].dt.dayofweek
df['dia_do_ano'] = df['game_date'].dt.dayofyear
```
